## ðŸ§ª 1. React KullanÄ±mÄ±nÄ±n MaaÅŸa Etkisi (BaÄŸÄ±msÄ±z t-testi)

```python
# React kullanÄ±mÄ±na gÃ¶re gruplarÄ± ayÄ±r
react_users = df[df['frontend_react'] == 1]['ortalama_maas']
non_react_users = df[df['frontend_react'] == 0]['ortalama_maas']

print(f"React kullananlar: {len(react_users)} kiÅŸi")
print(f"React kullanmayanlar: {len(non_react_users)} kiÅŸi")
print(f"\nReact kullananlarÄ±n ortalama maaÅŸÄ±: {react_users.mean():.2f} bin TL")
print(f"React kullanmayanlarÄ±n ortalama maaÅŸÄ±: {non_react_users.mean():.2f} bin TL")
print(f"Fark: {react_users.mean() - non_react_users.mean():.2f} bin TL")
```

```python
# BaÄŸÄ±msÄ±z t-testi uygula
t_stat, p_value = ttest_ind(react_users, non_react_users)

# Cohen's d hesapla
pooled_std = np.sqrt(((len(react_users) - 1) * react_users.var() + 
                     (len(non_react_users) - 1) * non_react_users.var()) / 
                    (len(react_users) + len(non_react_users) - 2))
cohens_d = (react_users.mean() - non_react_users.mean()) / pooled_std

# GÃ¼ven aralÄ±ÄŸÄ± hesapla
mean_diff = react_users.mean() - non_react_users.mean()
se_diff = np.sqrt(react_users.var()/len(react_users) + non_react_users.var()/len(non_react_users))
ci_lower = mean_diff - 1.96 * se_diff
ci_upper = mean_diff + 1.96 * se_diff

print("=== REACT KULLANIMI T-TEST SONUÃ‡LARI ===")
print(f"t-istatistiÄŸi: {t_stat:.4f}")
print(f"p-deÄŸeri: {p_value:.6f}")
print(f"Cohen's d: {cohens_d:.4f}")
print(f"%95 GÃ¼ven AralÄ±ÄŸÄ±: [{ci_lower:.2f}, {ci_upper:.2f}] bin TL")
print(f"\nYorum: React kullananlarÄ±n maaÅŸÄ± {'anlamlÄ± olarak yÃ¼ksek' if p_value < 0.05 else 'anlamlÄ± olarak farklÄ± deÄŸil'} (p {'<' if p_value < 0.05 else '>'} 0.05)")
```

```python
# GÃ¶rselleÅŸtirme
plt.figure(figsize=(10, 6))

# Box plot
plt.subplot(1, 2, 1)
data_to_plot = [react_users, non_react_users]
labels = ['React Kullananlar', 'React Kullanmayanlar']
plt.boxplot(data_to_plot, labels=labels)
plt.title('React KullanÄ±mÄ±na GÃ¶re MaaÅŸ DaÄŸÄ±lÄ±mÄ±')
plt.ylabel('MaaÅŸ (bin TL)')
plt.xticks(rotation=45)

# Histogram
plt.subplot(1, 2, 2)
plt.hist(react_users, alpha=0.7, label='React Kullananlar', bins=30)
plt.hist(non_react_users, alpha=0.7, label='React Kullanmayanlar', bins=30)
plt.title('React KullanÄ±mÄ±na GÃ¶re MaaÅŸ HistogramÄ±')
plt.xlabel('MaaÅŸ (bin TL)')
plt.ylabel('Frekans')
plt.legend()

plt.tight_layout()
plt.show()
```

## ðŸ¢ 2. Ã‡alÄ±ÅŸma Åžeklinin MaaÅŸa Etkisi (Tek YÃ¶nlÃ¼ ANOVA)

```python
# Ã‡alÄ±ÅŸma ÅŸekline gÃ¶re gruplarÄ± ayÄ±r
work_types = {}
work_type_names = {0: 'Remote', 1: 'Office', 2: 'Hybrid'}

for i in range(3):
    work_types[i] = df[df['calisma_sekli'] == i]['ortalama_maas']
    print(f"{work_type_names[i]}: {len(work_types[i])} kiÅŸi, Ortalama: {work_types[i].mean():.2f} bin TL")
```

```python
# ANOVA testi uygula
f_stat, p_value = f_oneway(work_types[0], work_types[1], work_types[2])

# Eta-squared hesapla
ss_between = sum(len(group) * ((group.mean() - df['ortalama_maas'].mean())**2) for group in work_types.values())
ss_total = sum((df['ortalama_maas'] - df['ortalama_maas'].mean())**2)
eta_squared = ss_between / ss_total

print("=== Ã‡ALIÅžMA ÅžEKLÄ° ANOVA SONUÃ‡LARI ===")
print(f"F-istatistiÄŸi: {f_stat:.4f}")
print(f"p-deÄŸeri: {p_value:.6f}")
print(f"Eta-squared: {eta_squared:.4f}")
print(f"\nYorum: Ã‡alÄ±ÅŸma ÅŸekli gruplarÄ± arasÄ±nda {'anlamlÄ± fark var' if p_value < 0.05 else 'anlamlÄ± fark yok'} (p {'<' if p_value < 0.05 else '>'} 0.05)")
```

```python
# Post-hoc Tukey HSD testi (eÄŸer ANOVA anlamlÄ±yasa)
if p_value < 0.05:
    print("\n=== POST-HOC TUKEY HSD TESTÄ° ===")
    
    # Veriyi hazÄ±rla
    tukey_data = []
    tukey_labels = []
    
    for i in range(3):
        tukey_data.extend(work_types[i])
        tukey_labels.extend([work_type_names[i]] * len(work_types[i]))
    
    # Tukey testi uygula
    tukey_result = pairwise_tukeyhsd(tukey_data, tukey_labels)
    print(tukey_result)
else:
    print("\nPost-hoc test gerekli deÄŸil (ANOVA anlamlÄ± deÄŸil)")
```

```python
# GÃ¶rselleÅŸtirme
plt.figure(figsize=(12, 5))

# Box plot
plt.subplot(1, 2, 1)
data_to_plot = [work_types[0], work_types[1], work_types[2]]
labels = ['Remote', 'Office', 'Hybrid']
plt.boxplot(data_to_plot, labels=labels)
plt.title('Ã‡alÄ±ÅŸma Åžekline GÃ¶re MaaÅŸ DaÄŸÄ±lÄ±mÄ±')
plt.ylabel('MaaÅŸ (bin TL)')

# Bar plot
plt.subplot(1, 2, 2)
means = [work_types[i].mean() for i in range(3)]
plt.bar(labels, means, color=['skyblue', 'lightgreen', 'lightcoral'])
plt.title('Ã‡alÄ±ÅŸma Åžekline GÃ¶re Ortalama MaaÅŸlar')
plt.ylabel('Ortalama MaaÅŸ (bin TL)')

for i, v in enumerate(means):
    plt.text(i, v + 1, f'{v:.1f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

## ðŸŒ 3. Åžirket Lokasyonunun MaaÅŸa Etkisi (Tek YÃ¶nlÃ¼ ANOVA)

```python
# Lokasyon gruplarÄ±nÄ± ayÄ±r (one-hot encoded sÃ¼tunlardan)
location_columns = [col for col in df.columns if col.startswith('location_')]
print(f"Lokasyon sÃ¼tunlarÄ±: {location_columns}")

location_groups = {}
for col in location_columns:
    location_name = col.replace('location_', '').replace('_', ' ').title()
    location_groups[location_name] = df[df[col] == 1]['ortalama_maas']
    print(f"{location_name}: {len(location_groups[location_name])} kiÅŸi, Ortalama: {location_groups[location_name].mean():.2f} bin TL")
```

```python
# ANOVA testi uygula
location_data = list(location_groups.values())
location_names = list(location_groups.keys())

f_stat, p_value = f_oneway(*location_data)

# Eta-squared hesapla
ss_between = sum(len(group) * ((group.mean() - df['ortalama_maas'].mean())**2) for group in location_data)
ss_total = sum((df['ortalama_maas'] - df['ortalama_maas'].mean())**2)
eta_squared = ss_between / ss_total

print("=== ÅžÄ°RKET LOKASYONU ANOVA SONUÃ‡LARI ===")
print(f"F-istatistiÄŸi: {f_stat:.4f}")
print(f"p-deÄŸeri: {p_value:.6f}")
print(f"Eta-squared: {eta_squared:.4f}")
print(f"\nYorum: Lokasyon gruplarÄ± arasÄ±nda {'anlamlÄ± fark var' if p_value < 0.05 else 'anlamlÄ± fark yok'} (p {'<' if p_value < 0.05 else '>'} 0.05)")
```

```python
# Post-hoc Tukey HSD testi (eÄŸer ANOVA anlamlÄ±ysa)
if p_value < 0.05:
    print("\n=== POST-HOC TUKEY HSD TESTÄ° ===")
    
    # Veriyi hazÄ±rla
    tukey_data = []
    tukey_labels = []
    
    for i, group in enumerate(location_data):
        tukey_data.extend(group)
        tukey_labels.extend([location_names[i]] * len(group))
    
    # Tukey testi uygula
    tukey_result = pairwise_tukeyhsd(tukey_data, tukey_labels)
    print(tukey_result)
else:
    print("\nPost-hoc test gerekli deÄŸil (ANOVA anlamlÄ± deÄŸil)")
```

```python
# GÃ¶rselleÅŸtirme
plt.figure(figsize=(15, 5))

# Box plot
plt.subplot(1, 2, 1)
plt.boxplot(location_data, labels=location_names)
plt.title('Åžirket Lokasyonuna GÃ¶re MaaÅŸ DaÄŸÄ±lÄ±mÄ±')
plt.ylabel('MaaÅŸ (bin TL)')
plt.xticks(rotation=45)

# Bar plot
plt.subplot(1, 2, 2)
means = [group.mean() for group in location_data]
plt.bar(location_names, means, color=['skyblue', 'lightgreen', 'lightcoral', 'gold'])
plt.title('Åžirket Lokasyonuna GÃ¶re Ortalama MaaÅŸlar')
plt.ylabel('Ortalama MaaÅŸ (bin TL)')
plt.xticks(rotation=45)

for i, v in enumerate(means):
    plt.text(i, v + 2, f'{v:.1f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

## ðŸ‘¥ 4. Cinsiyet BazlÄ± MaaÅŸ FarkÄ± (BaÄŸÄ±msÄ±z t-testi)

```python
# Cinsiyete gÃ¶re gruplarÄ± ayÄ±r
male_salary = df[df['cinsiyet'] == 0]['ortalama_maas']  # 0: Erkek
female_salary = df[df['cinsiyet'] == 1]['ortalama_maas']  # 1: KadÄ±n

print(f"Erkek geliÅŸtiriciler: {len(male_salary)} kiÅŸi")
print(f"KadÄ±n geliÅŸtiriciler: {len(female_salary)} kiÅŸi")
print(f"\nErkeklerin ortalama maaÅŸÄ±: {male_salary.mean():.2f} bin TL")
print(f"KadÄ±nlarÄ±n ortalama maaÅŸÄ±: {female_salary.mean():.2f} bin TL")
print(f"MaaÅŸ farkÄ±: {male_salary.mean() - female_salary.mean():.2f} bin TL")
print(f"Gender gap yÃ¼zdesi: {((male_salary.mean() - female_salary.mean()) / male_salary.mean() * 100):.2f}%")
```

```python
# BaÄŸÄ±msÄ±z t-testi uygula
t_stat, p_value = ttest_ind(male_salary, female_salary)

# Cohen's d hesapla
pooled_std = np.sqrt(((len(male_salary) - 1) * male_salary.var() + 
                     (len(female_salary) - 1) * female_salary.var()) / 
                    (len(male_salary) + len(female_salary) - 2))
cohens_d = (male_salary.mean() - female_salary.mean()) / pooled_std

# GÃ¼ven aralÄ±ÄŸÄ± hesapla
mean_diff = male_salary.mean() - female_salary.mean()
se_diff = np.sqrt(male_salary.var()/len(male_salary) + female_salary.var()/len(female_salary))
ci_lower = mean_diff - 1.96 * se_diff
ci_upper = mean_diff + 1.96 * se_diff

print("=== CÄ°NSÄ°YET BAZLI MAAÅž FARKI T-TEST SONUÃ‡LARI ===")
print(f"t-istatistiÄŸi: {t_stat:.4f}")
print(f"p-deÄŸeri: {p_value:.6f}")
print(f"Cohen's d: {cohens_d:.4f}")
print(f"%95 GÃ¼ven AralÄ±ÄŸÄ±: [{ci_lower:.2f}, {ci_upper:.2f}] bin TL")
print(f"Gender gap: {((male_salary.mean() - female_salary.mean()) / male_salary.mean() * 100):.2f}%")
print(f"\nYorum: Cinsiyetler arasÄ±nda {'anlamlÄ± maaÅŸ farkÄ± var' if p_value < 0.05 else 'anlamlÄ± maaÅŸ farkÄ± yok'} (p {'<' if p_value < 0.05 else '>'} 0.05)")
```

```python
# GÃ¶rselleÅŸtirme
plt.figure(figsize=(12, 5))

# Box plot
plt.subplot(1, 2, 1)
data_to_plot = [male_salary, female_salary]
labels = ['Erkek', 'KadÄ±n']
plt.boxplot(data_to_plot, labels=labels)
plt.title('Cinsiyete GÃ¶re MaaÅŸ DaÄŸÄ±lÄ±mÄ±')
plt.ylabel('MaaÅŸ (bin TL)')

# Bar plot
plt.subplot(1, 2, 2)
means = [male_salary.mean(), female_salary.mean()]
plt.bar(labels, means, color=['lightblue', 'lightpink'])
plt.title('Cinsiyete GÃ¶re Ortalama MaaÅŸlar')
plt.ylabel('Ortalama MaaÅŸ (bin TL)')

for i, v in enumerate(means):
    plt.text(i, v + 1, f'{v:.1f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

## ðŸ“ˆ 5. Deneyim-MaaÅŸ Korelasyonu

```python
# Kariyer seviyesi ile maaÅŸ arasÄ±ndaki korelasyonu hesapla
pearson_r, pearson_p = pearsonr(df['kariyer_seviyesi'], df['ortalama_maas'])
spearman_r, spearman_p = spearmanr(df['kariyer_seviyesi'], df['ortalama_maas'])

print("=== DENEYÄ°M-MAAÅž KORELASYONU ===")
print(f"Pearson korelasyonu: r = {pearson_r:.4f}, p = {pearson_p:.6f}")
print(f"Spearman korelasyonu: r = {spearman_r:.4f}, p = {spearman_p:.6f}")
print(f"\nYorum: Kariyer seviyesi ile maaÅŸ arasÄ±nda {'anlamlÄ± pozitif korelasyon var' if pearson_p < 0.05 else 'anlamlÄ± korelasyon yok'} (p {'<' if pearson_p < 0.05 else '>'} 0.05)")
```

```python
# GÃ¶rselleÅŸtirme
plt.figure(figsize=(12, 5))

# Scatter plot
plt.subplot(1, 2, 1)
plt.scatter(df['kariyer_seviyesi'], df['ortalama_maas'], alpha=0.6)
plt.xlabel('Kariyer Seviyesi')
plt.ylabel('MaaÅŸ (bin TL)')
plt.title('Kariyer Seviyesi vs MaaÅŸ Korelasyonu')

# Trend Ã§izgisi ekle
z = np.polyfit(df['kariyer_seviyesi'], df['ortalama_maas'], 1)
p = np.poly1d(z)
plt.plot(df['kariyer_seviyesi'], p(df['kariyer_seviyesi']), "r--", alpha=0.8)

# Box plot kariyer seviyesine gÃ¶re
plt.subplot(1, 2, 2)
career_levels = [df[df['kariyer_seviyesi'] == i]['ortalama_maas'] for i in range(1, 6)]
labels = ['Junior', 'Mid', 'Senior', 'Lead', 'Manager']
plt.boxplot(career_levels, labels=labels)
plt.title('Kariyer Seviyesine GÃ¶re MaaÅŸ DaÄŸÄ±lÄ±mÄ±')
plt.ylabel('MaaÅŸ (bin TL)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
